```python
import pandas as pd
import matplotlib.pyplot as plt 
plt.style.use("seaborn")
```

```python
df = pd.read_excel('RankingB1-30-08-2021 (1).xlsx')
df.head()
```

```python
#Padronizando nomes das colunas
df = df.rename(columns={"Distribuidora":"distribuidora", "UF": "uf", "Ranking":"ranking", "Tarifa Convencional":"tarifaConvencional", "Tarifa Branca - Ponta":"tarifaBrancaPonta", "Tarifa Branca - Intermediária": "tarifaBrancaIntermediaria", "Tarifa Branca - Fora ponta": "tarifaBrancaForaPonta","Resolução Homologatória": "resolHomolog","Início de vigência": "inicioVigencia"})
df.head(10)
```

```python
#Quais as 3 concessionárias com a tarifa convencional mais cara? 
df.nlargest(3, "tarifaConvencional")
```

```python
#Quais as 3 concessionárias com a tarifa convencional mais barata? 
df.nsmallest(3, "tarifaConvencional")
```

```python
#Quais as 5 concessionárias com a tarifa branca fora ponta mais baratas?
df.nsmallest(5,'tarifaBrancaForaPonta')
```

```python
#Quais as 5 concessionárias com a tarifa branca intermediaria mais baratas?
df.nsmallest(5,'tarifaBrancaIntermediaria')
```

```python
#Quais as 5 concessionárias com a tarifa branca ponta mais baratas?
df.nsmallest(5,'tarifaBrancaPonta')
```

```python
#Gráfico maiores tarifas convencionais

df.nlargest(10,'tarifaConvencional').plot.barh(x='distribuidora',y='tarifaConvencional')
plt.xlabel("Distribuidora")
plt.ylabel("Tarifa Convencional");
```

```python
#Plotar distribuidoras do nordeste
#Estados nordeste
nordeste = ['AL','BA','CE','MA','PB','PI','RN','SE']
dfnordeste = df[df['uf'].isin(nordeste)]
dfnordeste['distribuidora_uf'] = dfnordeste['distribuidora'] + '-' + dfnordeste['uf']
dfnordeste.head()
```
